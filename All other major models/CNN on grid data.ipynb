```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, f1_score
from sklearn.utils.class_weight import compute_class_weight
from scipy.interpolate import griddata
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# -------------------------------
# 1) Load dataset
# -------------------------------
df = pd.read_csv("phase_t1.0_U25.0-(0.1)_Mu60.0-(0.1).csv")

# -------------------------------
# 2) Physics-based labeling
# -------------------------------
def label_phase(rho, phi2, tol_rho=1e-3, tol_phi=1e-6):
    is_integer_rho = np.isclose(rho, np.round(rho), atol=tol_rho)
    is_zero_phi = phi2 < tol_phi
    return 0 if (is_integer_rho and is_zero_phi) else 1

df["Phase"] = df.apply(lambda row: label_phase(row["Density"], row["Phi2"]), axis=1)

# -------------------------------
# 3) Grid-preserving downsample
# -------------------------------
df["Ubin"] = pd.cut(df["Ubar"], bins=100)
df["Mubin"] = pd.cut(df["Mubar"], bins=200)
df_reduced = df.groupby(["Ubin","Mubin"]).apply(lambda g: g.sample(1, random_state=42)).reset_index(drop=True)
df = df_reduced.copy()

# -------------------------------
# 4) Prepare grid for CNN
# -------------------------------
U_vals = np.sort(df["Ubar"].unique())
mu_vals = np.sort(df["Mubar"].unique())
nU, nMu = len(U_vals), len(mu_vals)
print("Grid size (nU,nMu) =", nU, nMu)

features = ["Phi2", "Density"]
X_grid = np.zeros((nU, nMu, len(features)), dtype=np.float32)
y_phase_grid = np.zeros((nU, nMu), dtype=np.float32)
y_phi2_grid = np.zeros((nU, nMu), dtype=np.float32)
y_rho_grid = np.zeros((nU, nMu), dtype=np.float32)

for i, U in enumerate(U_vals):
    for j, mu in enumerate(mu_vals):
        row = df[(df["Ubar"]==U) & (df["Mubar"]==mu)]
        if not row.empty:
            X_grid[i,j,:] = row[features].values[0]
            y_phase_grid[i,j] = row["Phase"].values[0]
            y_phi2_grid[i,j] = row["Phi2"].values[0]
            y_rho_grid[i,j] = row["Density"].values[0]

# Add batch + channel dims
X = np.expand_dims(X_grid, axis=0)
y_phase = np.expand_dims(y_phase_grid, axis=(0,-1))
y_phi2  = np.expand_dims(y_phi2_grid, axis=(0,-1))
y_rho   = np.expand_dims(y_rho_grid, axis=(0,-1))

# -------------------------------
# 5) Compute class weights
# -------------------------------
y_flat = y_phase_grid.ravel().astype(int)
classes = np.unique(y_flat)
cw = compute_class_weight('balanced', classes=classes, y=y_flat)
class_weights = {int(c): float(w) for c,w in zip(classes, cw)}
print("Computed class_weights:", class_weights)

# -------------------------------
# 6) Custom loss with label smoothing
# -------------------------------
def binary_crossentropy_with_smoothing(y_true, y_pred, label_smoothing=0.1):
    y_true = tf.cast(y_true, tf.float32)
    y_true = y_true * (1 - label_smoothing) + 0.5 * label_smoothing
    return tf.keras.losses.binary_crossentropy(y_true, y_pred)

# -------------------------------
# 7) Build multi-task CNN
# -------------------------------
inputs = keras.Input(shape=X.shape[1:])
x = layers.Conv2D(32, (3,3), padding="same", activation="relu")(inputs)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.2)(x)
x = layers.Conv2D(64, (3,3), padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.2)(x)
x = layers.Conv2D(128, (3,3), padding="same", activation="relu")(x)

phase_out = layers.Conv2D(1, (1,1), activation="sigmoid", name="phase")(x)
phi2_out  = layers.Conv2D(1, (1,1), activation="linear", name="phi2")(x)
rho_out   = layers.Conv2D(1, (1,1), activation="linear", name="rho")(x)

model = keras.Model(inputs=inputs, outputs=[phase_out, phi2_out, rho_out])
model.compile(
    optimizer="adam",
    loss={
        "phase": lambda y_true, y_pred: binary_crossentropy_with_smoothing(y_true, y_pred, label_smoothing=0.1),
        "phi2": "mse",
        "rho": "mse"
    },
    loss_weights={"phase":1.0, "phi2":0.3, "rho":0.3},
    metrics={"phase":"accuracy"}
)
model.summary()

# -------------------------------
# 8) Sample weights
# -------------------------------
sw_phase = np.where(y_phase==0, class_weights.get(0,1.0), class_weights.get(1,1.0)).astype(np.float32)
sw_phi2  = np.ones_like(y_phi2, dtype=np.float32)
sw_rho   = np.ones_like(y_rho, dtype=np.float32)
sample_weights = [sw_phase, sw_phi2, sw_rho]

# -------------------------------
# 9) Train model
# -------------------------------
history = model.fit(
    X,
    [y_phase, y_phi2, y_rho],
    epochs=60,
    batch_size=1,
    verbose=2,
    sample_weight=sample_weights
)

# -------------------------------
# 10) Predict on fixed training grid
# -------------------------------
y_pred_phase, y_pred_phi2, y_pred_rho = model.predict(X)
y_pred_phase = y_pred_phase[0,:,:,0]
y_pred_phi2  = y_pred_phi2[0,:,:,0]
y_pred_rho   = y_pred_rho[0,:,:,0]

# -------------------------------
# 11) Zoom & evaluate on subset
# -------------------------------
def zoom_evaluate(U_min=None, U_max=None, mu_min=None, mu_max=None):
    # Keep mask on original grid
    U_mask = np.ones_like(U_vals, dtype=bool) if U_min is None else (U_vals>=U_min) & (U_vals<=U_max)
    mu_mask = np.ones_like(mu_vals, dtype=bool) if mu_min is None else (mu_vals>=mu_min) & (mu_vals<=mu_max)

    y_pred_phase_sub = y_pred_phase[U_mask][:, mu_mask]
    UU_sub, MM_sub = np.meshgrid(U_vals[U_mask], mu_vals[mu_mask], indexing="ij")

    # Interpolate true labels on the same subgrid
    phase_true_grid = griddata(df[["Ubar","Mubar"]].values,
                               df["Phase"].values,
                               (UU_sub, MM_sub),
                               method="nearest")
    y_true = phase_true_grid.ravel().astype(int)
    probs = y_pred_phase_sub.ravel()

    # Tune threshold
    best_thr, best_f1 = 0.5, 0.0
    for thr in np.linspace(0.1, 0.9, 33):
        preds = (probs > thr).astype(int)
        f1 = f1_score(y_true, preds, zero_division=0)
        if f1 > best_f1:
            best_f1, best_thr = f1, thr
    print(f"Best threshold found: {best_thr:.3f}  (F1 = {best_f1:.4f})")

    phase_pred_class = (y_pred_phase_sub > best_thr).astype(int)

    # Report
    print("\n📊 Classification Report (CNN on zoomed grid):")
    print(classification_report(y_true, phase_pred_class.ravel(), digits=4, zero_division=0))

    # Plot
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.contourf(U_vals[U_mask], mu_vals[mu_mask], y_pred_phase_sub.T, 50, cmap="coolwarm")
    plt.colorbar(label="Phase probability")
    plt.xlabel("Ū"); plt.ylabel("μ̄"); plt.title("CNN Phase Probability")

    plt.subplot(1,2,2)
    plt.contourf(U_vals[U_mask], mu_vals[mu_mask], phase_pred_class.T, levels=[0,0.5,1.0], cmap="coolwarm")
    plt.colorbar(label="Phase class")
    plt.xlabel("Ū"); plt.ylabel("μ̄"); plt.title("CNN Phase Classification (thresholded)")
    plt.tight_layout(); plt.show()
def zoom_evaluate_extrap(U_min=None, U_max=None, mu_min=None, mu_max=None, nU_new=60, nMu_new=60):
    # Step 1: define target grid
    U_min = U_vals.min() if U_min is None else U_min
    U_max = U_vals.max() if U_max is None else U_max
    mu_min = mu_vals.min() if mu_min is None else mu_min
    mu_max = mu_vals.max() if mu_max is None else mu_max

    UU_new = np.linspace(U_min, U_max, nU_new)
    MM_new = np.linspace(mu_min, mu_max, nMu_new)
    UU_grid, MM_grid = np.meshgrid(UU_new, MM_new, indexing="ij")

    # Step 2: interpolate original features onto this grid
    Phi2_grid = griddata(df[["Ubar","Mubar"]].values, df["Phi2"].values, (UU_grid, MM_grid), method="nearest")
    Rho_grid  = griddata(df[["Ubar","Mubar"]].values, df["Density"].values, (UU_grid, MM_grid), method="nearest")
    X_new = np.stack([Phi2_grid, Rho_grid], axis=-1)
    X_new = np.expand_dims(X_new, axis=0)  # batch dimension

    # Step 3: resize/interpolate to CNN input shape
    from tensorflow.image import resize
    X_cnn_input = tf.image.resize(X_new, (X.shape[1], X.shape[2])).numpy()

    # # Step 4: predict
    # y_pred_phase, y_pred_phi2, y_pred_rho = model.predict(X_cnn_input)
    # y_pred_phase = tf.image.resize(y_pred_phase[0], (nU_new, nMu_new)).numpy()
    # y_pred_phi2  = tf.image.resize(y_pred_phi2[0],  (nU_new, nMu_new)).numpy()
    # y_pred_rho   = tf.image.resize(y_pred_rho[0],   (nU_new, nMu_new)).numpy()

    # Step 4: predict
    y_pred_phase, y_pred_phi2, y_pred_rho = model.predict(X_cnn_input)
    
    y_pred_phase = tf.image.resize(y_pred_phase[0], (nU_new, nMu_new)).numpy()[..., 0]
    y_pred_phi2  = tf.image.resize(y_pred_phi2[0],  (nU_new, nMu_new)).numpy()[..., 0]
    y_pred_rho   = tf.image.resize(y_pred_rho[0],   (nU_new, nMu_new)).numpy()[..., 0]


    # Step 5: compute true labels (nearest from original data)
    phase_true_grid = griddata(df[["Ubar","Mubar"]].values, df["Phase"].values, (UU_grid, MM_grid), method="nearest")
    y_true = phase_true_grid.ravel().astype(int)
    probs = y_pred_phase.ravel()

    # Step 6: tune threshold
    best_thr, best_f1 = 0.5, 0.0
    for thr in np.linspace(0.1, 0.9, 33):
        preds = (probs > thr).astype(int)
        f1 = f1_score(y_true, preds, zero_division=0)
        if f1 > best_f1:
            best_f1, best_thr = f1, thr
    print(f"Best threshold: {best_thr:.3f}  (F1 = {best_f1:.4f})")
    phase_pred_class = (y_pred_phase > best_thr).astype(int)

    # Step 7: plot
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.contourf(UU_new, MM_new, y_pred_phase.T, 50, cmap="coolwarm")
    plt.colorbar(label="Phase probability")
    plt.xlabel("Ū"); plt.ylabel("μ̄"); plt.title("CNN Phase Probability")

    plt.subplot(1,2,2)
    plt.contourf(UU_new, MM_new, phase_pred_class.T, levels=[0,0.5,1.0], cmap="coolwarm")
    plt.colorbar(label="Phase class")
    plt.xlabel("Ū"); plt.ylabel("μ̄"); plt.title("CNN Phase Classification (thresholded)")
    plt.tight_layout(); plt.show()

# -------------------------------
# 12) Example usage
# -------------------------------
# Full grid
zoom_evaluate()

# Zoomed region
zoom_evaluate_extrap(U_min=0, U_max=30, mu_min=0, mu_max=65)
```
